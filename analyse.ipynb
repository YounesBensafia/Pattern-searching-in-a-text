```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

len_values = []
time_array = []
time_tree = []
time_fm = []

with open('./csv/results.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader:
        len_values.append(float(row[0]))
        time_array.append(float(row[1]))
        time_tree.append(float(row[2]))  
        time_fm.append(float(row[3]))    

len_values_np = np.array(len_values)
time_array_np = np.array(time_array)
time_tree_np = np.array(time_tree)
time_fm_np = np.array(time_fm)

unique_len_values_np, unique_indices = np.unique(len_values_np, return_index=True)
time_array_np = time_array_np[unique_indices]
time_tree_np = time_tree_np[unique_indices]
time_fm_np = time_fm_np[unique_indices]

len_values_1 = np.linspace(unique_len_values_np.min(), 8000, 80) 
time_array_1 = make_interp_spline(unique_len_values_np, time_array_np)(len_values_1)
time_tree_1 = make_interp_spline(unique_len_values_np, time_tree_np)(len_values_1)
time_fm_1 = make_interp_spline(unique_len_values_np, time_fm_np)(len_values_1)

plt.figure(figsize=(20, 10))

plt.plot(len_values_1, time_array_1, label='Time Array', color='b', marker='o', linewidth=1)
plt.plot(len_values_1, time_tree_1, label='Time Tree', color='r', marker='x', linewidth=1)
plt.plot(len_values_1, time_fm_1, label='Time FM', color='g', marker='^', linewidth=1)

plt.xlabel('Length (len)')
plt.ylabel('Time (in seconds or units)')
plt.title('Comparison of Time vs Length')

plt.legend()

plt.show()
```

```python

len_values_1_4 = np.linspace(6000,6440, 1000000) 
time_tree_1_4 = make_interp_spline(unique_len_values_np, time_tree_np, k=3)(len_values_1_4)
time_fm_1_4 = make_interp_spline(unique_len_values_np, time_fm_np, k=3)(len_values_1_4)

time_tree_1_4 = np.clip(time_tree_1_4, 0, None)  
time_fm_1_4 = np.clip(time_fm_1_4, 0, None) 

plt.figure(figsize=(20, 10))

plt.plot(len_values_1_4, time_tree_1_4, label='Time Tree', color='r', linewidth=3)
plt.plot(len_values_1_4, time_fm_1_4, label='Time FM', color='g', linewidth=3)

plt.xlabel('Length (len)')
plt.ylabel('Time (in seconds or units)')
plt.title('Comparison of Time vs Length for Tree and FM with High 1ness')

plt.legend()

plt.show()

# ======================================================= ICI J'AI ZOOME ===============================================
```
